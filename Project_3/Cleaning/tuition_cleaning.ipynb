```python
# import dependencies
import pandas as pd 
```

```python
# load in file
tuition_file = "../Resources/tuition_trends.csv"
expense_file = "../Resources/athletic_expense_report.csv"
```

```python
# turn file into pandas df
expense_data = pd.read_csv(expense_file)
tuition_data = pd.read_csv(tuition_file)


# expense_data.head()
tuition_data.head()
```

```python
tuition_data.columns
```

```python
# recreating football schools list 
schools_wanted = expense_data['Data'].unique()
football_schools = list(schools_wanted)
len(schools_wanted)
# print(football_schools)
```

```python
# dataframe of schools that are in football list
tuition_df = tuition_data[tuition_data['Unnamed: 1'].isin(football_schools)]
tuition_df
```

```python
# getting list of schools that are in the tuition dataframe
tuition_list = list(tuition_df['Unnamed: 1'].unique())
# tuition_list
```

```python
# shows what schools from football list are missing 
missing_schools = list(set(football_schools) - set(tuition_list))
missing_schools
```

```python
# dropping unwanted columns and chaning column names 
tuition_data = tuition_df.drop('Unnamed: 0', axis=1)
tuition_data = tuition_df.rename(columns={'Unnamed: 1':'School', 'Unnamed: 2':'Living_Type'})
tuition_data
```

```python
tuition_schools = list(tuition_data['School'].unique())
len(tuition_schools)
```

```python
# saving new_csv
tuition_data.to_csv('../Cleaned_Resources/tuition_data.csv')
```
